### Starting the Calculation

First we need to define the working directory. Everything we need will be copied from the examplex on the cluster into the working directory.

```python
RFDIFFUSION_WORKING_DIR = Path.cwd() / "protein_design_w_RFDiffusion"
```

Now we need to write information on the example. Looking at the examples in the RFDiffusion folder gives more options on what can be done.

```python
run_file = f"""
#!/bin/bash

# Load software module 
module load bio/rfdiffusion/1.1.0

# Create directories in your working directory
mkdir -p {RFDIFFUSION_WORKING_DIR}/inputs
mkdir -p {RFDIFFUSION_WORKING_DIR}/outputs
mkdir -p {RFDIFFUSION_WORKING_DIR}/schedules

# Copy input PDB file to your working directory
cp $RFDIFFUSION_HOME/examples/input_pdbs/5TPN.pdb {RFDIFFUSION_WORKING_DIR}/inputs/

# Run RFDiffusion 
HYDRA_FULL_ERROR=1  $RFDIFFUSION_HOME/scripts/run_inference.py \\
  inference.input_pdb={RFDIFFUSION_WORKING_DIR}/inputs/5TPN.pdb \\
  inference.output_prefix={RFDIFFUSION_WORKING_DIR}/outputs/motifscaffolding \\
  inference.schedule_directory_path={RFDIFFUSION_WORKING_DIR}/schedules \\
  inference.num_designs=3 \\
  'contigmap.contigs=[10-40/A163-181/10-40]'
"""

RUN_PATH = RFDIFFUSION_WORKING_DIR / "run.sh"  # file name!

with open(RUN_PATH, "w") as file:
    file.write(run_file)
```

Now we start the calculation by executing the cell below:

```python
os.system(f'echo "Running file {RUN_PATH}"')
os.system(f"bash {RUN_PATH}")
```
